```python
import os
import topogenesis as tg
import pyvista as pv
import trimesh as tm
import numpy as np
import pickle
import csv
```

```python
#This script is to determine which voxels need facades and different floor types in revit
########################################################################################
#Loading the complete lattice
lattice_path = os.path.relpath("data\lattice_complete_3_6.csv")
envelope_lattice = tg.lattice_from_csv(lattice_path)
coords = []

#Creating a [x, y, z] coordinates list
for i in range(len(envelope_lattice)):
    for j in range(len(envelope_lattice[0])):
        for k in range(len(envelope_lattice[0][0])):
            if  envelope_lattice[i][j][k] == True:
                coords.append([i, j, k])

#Finding neighbours and saving where none are found
calc = [-1, 1]
result_list = []
for i in coords:
    x = i[0]
    y = i[1]
    z = i[2]
    temp_list = [0,0,0,0]
    z_list =[0]
    for j in calc:
        if [x+j, y, z] not in coords:
            if j == -1:
                temp_list[0] = 1
            else:
                temp_list[2] = 1
        if [x, y+j, z] not in coords:
            if j == -1:
                temp_list[3] = 1
            else:
                temp_list[1] = 1
    if [x, y, z+1] not in coords:
        z_list[0] = 1
    result_list.append([x,y,z,temp_list,z_list])

#Pickle the data for use in revit_pipeline.gh
with open("data\facade_3_6.txt", "wb") as fp:   #Pickling
    pickle.dump(result_list, fp, protocol=2)
########################################################################################
print("done")
    
```
